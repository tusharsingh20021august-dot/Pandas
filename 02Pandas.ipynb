```python
# Joining data

cus_1 = pd.DataFrame({'cust_id': [101,102,],
                         'order_amt':[200,250]})

cus_2 = pd.DataFrame({'cust_id': [103,104,],
                         'order_amt':[300,350]})
```

```python
# Joining data with Concatenate

concat_1 = pd.concat([cus_1,cus_2], axis =0, ignore_index= True)
concat_1
```
